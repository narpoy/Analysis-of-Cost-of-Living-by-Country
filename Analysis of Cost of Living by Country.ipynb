# Analysis of Cost of Living by Country

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
data=pd.read_csv(r'Cost_of_Living_Index_by_Country_2024.csv')
```

```python
print(data.head())
```

```python
data_sorted = data.sort_values(by='Cost of Living Index', ascending=False)
```

```python
top_10_expensive = data_sorted.head(10)
print("Top 10 Most Expensive Countries:")
print(top_10_expensive[['Country', 'Cost of Living Index']])
```

```python
top_10_cheapest = data_sorted.tail(10)
print("Top 10 Cheapest Countries:")
print(top_10_cheapest[['Country', 'Cost of Living Index']])
```

# Visualizing Data

```python
plt.figure(figsize=(14, 8))

plt.subplot(1, 2, 1)
sns.barplot(data=top_10_expensive, x='Cost of Living Index', y='Country', palette='viridis')
plt.title('Top 10 Most Expensive Countries')
plt.xlabel('Cost of Living Index')

plt.subplot(1, 2, 2)
sns.barplot(data=top_10_cheapest, x='Cost of Living Index', y='Country', palette='viridis')
plt.title('Top 10 Cheapest Countries')
plt.xlabel('Cost of Living Index')

plt.tight_layout()
plt.show()
```

```python
categories = ['Groceries Index', 'Restaurant Price Index', 'Rent Index']
```

```python
for category in categories:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=top_10_expensive, x=category, y='Country', palette='viridis')
    plt.title(f'{category} for the Top 10 Most Expensive Countries')
    plt.xlabel(category)
    plt.show()
```

```python
for category in categories:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=top_10_cheapest, x=category, y='Country', palette='viridis')
    plt.title(f'{category} for the Top 10 Cheapest Countries')
    plt.xlabel(category)
    plt.show()
```
